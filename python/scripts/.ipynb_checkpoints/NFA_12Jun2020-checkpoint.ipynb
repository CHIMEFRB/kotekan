```python
import numpy as np
import threading
import socket
import struct
import requests
import time
import matplotlib.pyplot as plt
import serial
%matplotlib inline
```

```python
def sample_spectrum(samples=1024):
    url = 'http://localhost:12048/spectrum_output/xengine/spectrum'
    payload = '{"integration_length":%d}'%(samples)
    headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8'}
    r = requests.post(url, data=payload, headers=headers)
    return r.json()

def set_freq(freq):
    url = 'http://localhost:12048/airspy_input/freq'
    payload = '{"freq":%d}' %(freq)
    headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8'}
    r = requests.post(url, data=payload, headers=headers)
    return r

def set_gain(lna=10,mix=10,ifa=10):
    url = 'http://localhost:12048/airspy_input/freq'
    payload = '{"gain_lna":%d, "gain_mix":%d, "gain_if":%d}' %(lna,mix,ifa)
    headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8'}
    r = requests.post(url, data=payload, headers=headers)
    return r.json()    

def dB(d):
    return 10*np.log10(d)

def init_ns():
    return serial.Serial('/dev/tty.usbmodem14601', 115200)

def close_ns(s):
    s.close()
    return

def toggle_ns(s,powered=False):
    if powered:
        s.write('on\r'.encode())
    else:
        s.write('off\r'.encode())
    return
```

```python
from tqdm.notebook import tqdm

def get_spectrum(freqlist,gainlist=[]):
    specs = np.empty((0,1024))
    ifa = 10
    gl=gainlist
    pbar = tqdm(freqlist,ncols=800)
    for i,f in enumerate(pbar):
        set_freq(f)
        time.sleep(0.1)
        if (len(gainlist)==len(freqlist)): #use given gains
            ifa = gainlist[i]
            r = set_gain(ifa=ifa,mix=ifa,lna=ifa)
            pbar.set_description("Processing %d MHz (IF gain %d)" % (f,ifa))
        else: #find our own gains & return
            r = set_gain(ifa=ifa,mix=ifa,lna=ifa)
            pbar.set_description("Processing %d MHz (IF gain %d)" % (f,ifa))
            while (r['rms'] < 300):
                ifa+=1
                r = set_gain(ifa=ifa,mix=ifa,lna=ifa)
                pbar.set_description("Processing %d MHz (IF gain %d)" % (f,ifa))
            while (r['railfrac'] > 0):
                ifa-=1
                r = set_gain(ifa=ifa,mix=ifa,lna=ifa)
                pbar.set_description("Processing %d MHz (IF gain %d)" % (f,ifa))
        gl = np.append(gl,ifa)
        sp = np.array(sample_spectrum()['stokes_i'])
        specs = np.append(specs,sp[np.newaxis,:],axis=0)
    pbar.set_description("Spectrum complete!")
    return specs,gl
```

```python
def y_fac(on,off):
    return on/off

def y_fac_avg(on,off):
    edge=128
    y=y_fac(on,off)
    nf = y.shape[0]
    y_avg = np.zeros(nf)
    for i in np.arange(nf):
        med = np.median(y[i,edge:-edge])
        rms = np.std(y[i,edge:-edge])
        q = np.array(np.where((y[i,edge:-edge] < med+3*rms) * (y[i,edge:-edge] > med-3*rms))[0])
        y_avg[i] = np.median(y[i,edge+q])
    return y_avg

def y2nf(f,y):
    enr_raw = np.array([[0.01,0.10,1.0,2.0,3.0], [5.20, 5.36, 5.19, 5.01, 4.85]])
    enr = np.interp(f,enr_raw[0,:]*1e6,enr_raw[1,:])
    return 10*np.log10( (10**(enr/10) / (y-1)) )

def nf2tsys(nf):
    tamb=290
    return tamb*(10**(nf/10) -1)
```

```python
from scipy.signal import convolve, boxcar

def smooth(x,len=5):
    halflen=int(len/2)
    y = convolve(x,boxcar(len))[halflen:-halflen]/len
    return y
```

```python
def measure_yfac(freqlist):
    input("Turn on the noise source and hit enter to continue!")
    #noise source on, gains free
    s,g = get_spectrum(freqlist)
    noise_on = s
    input("Turn off the noise source and hit enter to continue!")
    #noise source off, reuse gain settings
    s,g2 = get_spectrum(freqlist,gainlist=g)
    noise_off = s
    ya = y_fac_avg(noise_on,noise_off)
    return noise_on,noise_off,ya

def auto_measure_yfac(freqlist):
    specs = np.empty((0,1024,2))
    sp = np.empty((1024,2))
    ifa = 10
    s = init_ns()
    pbar = tqdm(freqlist,ncols=800)
    for i,f in enumerate(pbar):
        set_freq(f)
        for ns_power in (1,0):
            toggle_ns(s,powered=ns_power)
            time.sleep(0.1)
            if (ns_power):
                r = set_gain(ifa=ifa,mix=ifa,lna=ifa)
                pbar.set_description("Processing %d MHz (IF gain %d)" % (f,ifa))
                while (r['rms'] < 300):
                    ifa+=1
                    r = set_gain(ifa=ifa,mix=ifa,lna=ifa)
                    pbar.set_description("Processing %d MHz (IF gain %d)" % (f,ifa))
                while (r['railfrac'] > 0):
                    ifa-=1
                    r = set_gain(ifa=ifa,mix=ifa,lna=ifa)
                    pbar.set_description("Processing %d MHz (IF gain %d)" % (f,ifa))
            sp[:,ns_power] = np.array(sample_spectrum()['stokes_i'])
        specs = np.append(specs,sp[np.newaxis,:,:],axis=0)
    pbar.set_description("Spectrum complete!")
    s.close()
    return specs
```

```python
shfreqlist = np.arange(1501,300,-50)
specs = auto_measure_yfac(shfreqlist)
ya = y_fac_avg(specs[:,:,1],specs[:,:,0])
```

```python
plt.figure()
plt.plot(shfreqlist,y2nf(shfreqlist,ya),'.')
plt.ylim(0,1)
plt.xlabel('Frequency (MHz)')
plt.ylabel('Noise Figure (dB)')

plt.figure()
plt.plot(shfreqlist,nf2tsys(y2nf(shfreqlist,ya)),'.')
plt.ylim(0,100)
plt.xlabel('Frequency (MHz)')
plt.ylabel('Tsys (K)')
```
